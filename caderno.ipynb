```python
# Condicionais aninhadas

print("Bem-vindo(a) à montanha russa!")
altura = int(input("Qual é sua altura em centímetros?\n-> "))

if altura >= 120:
    print("Você pode andar na montanha russa!")
    ingresso = 0
    idade = int(input("Qual é sua idade?\n-> "))
    if idade >= 18:
        print("Seu ingresso custará R$ 15,00.")
        ingresso += 15
    elif 18 > idade >= 10:
        print("Seu ingresso custará R$ 10,00.")
        ingresso += 10
    elif 55 >= idade >= 45:
        print("Você pode entrar de graça!")
        ingresso = 0
    else:
        print("Seu ingresso custará R$ 5,00.")
        ingresso += 5
    comprar_foto = input("Você gostaria de pagar R$ 1,00 a mais e ganhar uma foto? (S ou N)")
    if comprar_foto == "S":
        ingresso += 1
        print(f"Seu total a pagar é R$ {ingresso}")
    else:
        print(f"Seu total a pagar é R$ {ingresso}")
else:
    print("Sinto lhe dizer, mas você está abaixo da altura mínima necessária para andar na montanha russa.")
```

```python
# Calculadora de IMC com categorias de peso

peso = float(input("Qual é seu peso em quilos?\n-> "))
altura = float(input("Qual é sua altura em metros?\n-> "))

imc = (peso / (altura ** 2))
if imc < 18.5:
    print("Seu IMC é:", round(imc, 1), "\nVocê atualmente está com baixo peso.")
elif 18.5 <= imc < 25:
    print("Seu IMC é:", round(imc, 1), "\nSeu peso é regular.")
else:
    print("Seu IMC é:", round(imc, 1), "\Você atualmente está com sobrepeso.")
```

```python
# sum() function

goals_per_season = [15, 14, 16, 18, 19, 21, 26, 28, 31]
career_goals = sum(goals_per_season)

print(career_goals)

assists_per_season = [5, 3, 7, 8, 10, 12, 15, 12, 13]
career_assists = 0

# sum() function done manually

for season in assists_per_season:
    career_assists += season

print(career_assists)
```

```python
# max() and min() functions

goals_per_season = [15, 14, 16, 18, 19, 21, 26, 28, 31]
assists_per_season = [5, 3, 7, 8, 10, 12, 15, 12, 13]

print(max(goals_per_season), "-", min(goals_per_season))
print(max(assists_per_season), "-", min(assists_per_season), "\n")

# max() function done manually

max_goals = goals_per_season[0]
max_assists = assists_per_season[0]

for season in goals_per_season:
    if season > max_goals:
        max_goals = season

for season in assists_per_season:
    if season > max_assists:
        max_assists = season
        
# min() function done manuallly

min_goals = goals_per_season[0]
min_assists = assists_per_season[0]

for season in goals_per_season:
    if season < min_goals:
        min_goals = season

for season in assists_per_season:
    if season < min_assists:
        min_assists = season

# Checking the results

print(max_goals, "-", min_goals)
print(max_assists, "-", min_assists)
```

```python
# "Love Score" Calculator

def calculate_love_score(name1, name2):
    
    combined_names = (name1 + name2).upper()
    
    true_counter = 0
    love_counter = 0
    
    for letter in combined_names:
        if letter in "TRUE":
            true_counter += 1
        if letter in "LOVE":
            love_counter += 1
    
    true_counter = str(true_counter)
    love_counter = str(love_counter)
    love_score = true_counter + love_counter
    
    print(f"Your love score is {love_score}")

calculate_love_score("Angela", "Jack")
```

## Glossário

### A

### C

Cadeia (de caracteres) -> string

Cadeia formatada -> f-string

### D

Depurar -> debug

### E

Entrada -> input

Erro -> bug

### F

Falha -> bug

Flutuante (vírgula flutuante, número real) -> floating point, float

### L

Laço -> loop

### O

Operador de atribuição -> assignment operator

### R

Ramo -> branch (Git)

### S

Subtraço -> underline, underscore

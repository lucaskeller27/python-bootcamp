```python
# Calculadora de IMC com categorias de peso

peso = float(input("Qual é seu peso em quilos?\n-> "))
altura = float(input("Qual é sua altura em metros?\n-> "))

imc = (peso / (altura ** 2))
if imc < 18.5:
    print("Seu IMC é:", round(imc, 1), "\nVocê atualmente está com baixo peso.")
elif 18.5 <= imc < 25:
    print("Seu IMC é:", round(imc, 1), "\nSeu peso é regular.")
else:
    print("Seu IMC é:", round(imc, 1), "\Você atualmente está com sobrepeso.")
```

```python
# Função sum()

gols_por_temporada = [15, 14, 16, 18, 19, 21, 26, 28, 31]
gols_na_carreira = sum(gols_por_temporada)

print(gols_na_carreira)

assistencias_por_temporada = [5, 3, 7, 8, 10, 12, 15, 12, 13]
assistencias_na_carreira = 0

# Função sum() feita manualmente

for temporada in assistencias_por_temporada:
    assistencias_na_carreira += temporada

print(assistencias_na_carreira)
```

```python
# Funções max() e min()

gols_por_temporada = [15, 14, 16, 18, 19, 21, 26, 28, 31]
assistencias_por_temporada = [5, 3, 7, 8, 10, 12, 15, 12, 13]

print(f"{max(gols_por_temporada)} - {min(gols_por_temporada)}")
print(f"{max(assistencias_por_temporada)} - {min(assistencias_por_temporada)}\n")

# Função max() feita manualmente

maximo_gols = gols_por_temporada[0]
maximo_assistencias = assistencias_por_temporada[0]

for temporada in gols_por_temporada:
    if temporada > maximo_gols:
        maximo_gols = temporada

for temporada in assistencias_por_temporada:
    if temporada > maximo_assistencias:
        maximo_assistencias = temporada
        
# Função min() feita manualmente

minimo_gols = gols_por_temporada[0]
minimo_assistencias = assistencias_por_temporada[0]

for temporada in gols_por_temporada:
    if temporada < minimo_gols:
        minimo_gols = temporada

for temporada in assistencias_por_temporada:
    if temporada < minimo_assistencias:
        minimo_assistencias = temporada

# Conferir os resultados

print(f"{maximo_gols} - {minimo_gols}")
print(f"{maximo_assistencias} - {minimo_assistencias}")
```

#### Dia 8 - Parâmetros e argumentos de funções

```python
# Dar entrada a funções

def cumprimentar(nome): # Parâmetro
    print(f"Olá {nome}, bom dia!")

cumprimentar("Lucas") # Argumento
```

```python
# Calculadora de semanas restantes de vida

def vida_em_semanas(idade):
    semanas_restantes = (90 - int(idade)) * 52
    print(f"Caso viva até os 90 anos de idade, você tem {semanas_restantes} semanas restantes de vida.")
```

```python
# Calculadora de semanas restantes de vida (com idade de morte personalizável)

def vida_em_semanas(idade_atual, idade_de_morte):
    semanas_restantes = (int(idade_de_morte) - int(idade_atual)) * 52
    print(f"Caso viva até os {idade_de_morte} anos de idade, você tem {semanas_restantes} semanas restantes de vida.")

vida_em_semanas(20, 91)
```

```python
# Funções com mais de uma entrada

def cumprimentar(nome, local):
    print(f"Bom dia {nome}, como está o tempo agora em {local}?")

cumprimentar("Lucas", "Desterro")

# Argumentos posicionais e nomeados

cumprimentar("Porto Alegre", "Joaquim")
cumprimentar(local="Porto Alegre", nome="Joaquim")
```

```python
# Calculadora de Amor Puro

def calcular_amor(nome1, nome2):
    
    # Definir variáveis
    nomes = (nome1 + nome2).upper()
    contador_amor = 0
    contador_puro = 0
    
    # Contar as letras
    for letra in nomes:
        if letra in "AMOR":
            contador_amor += 1
        if letra in "PURO":
            contador_puro += 1
    
    # Calcular Amor Puro
    amor_puro = str(contador_amor) + str(contador_puro)
    print(f"A pontuação de Amor Puro do casal é {amor_puro}.")

calcular_amor("Eduardo", "Monica")
```
